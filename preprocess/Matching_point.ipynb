```python
def matching_point(mentors, mentee):
    scores = []
    for index, mentor in mentors.iterrows():
        #skills first
        score = 0
        skill_match = 0
        count_skill = sum(map(lambda x : x == 1, mentee.skill_list[0]))
        for skill in range(len(mentor.skill_list)):
            if((mentor.skill_list[skill] == mentee.skill_list[0][skill]) and mentee.skill_list[0][skill] == 1):
                skill_match = skill_match +1
        score = skill_match/count_skill *3.5
        
        #company score
        if(mentee.cur_company[0] == mentor.cur_company):
            score = score + 0.5
            
         #location score
        if(mentee.location[0] == mentor.location):
            score = score + 0.5
            
        #school_index score
        school_match =0
        for sch in mentee.school_index[0]:
            if(sch in mentor.school_index):
                school_match = school_match + 1
        score = score + school_match/len(mentee.school_index[0])*0.5
            
        scores.append(score)  
    mentors_temp = mentors
    mentors_temp['matching_point'] = scores
    return mentors_temp
```

```python
def mentee_setting():
    mentee = pd.DataFrame(columns = {'skill_list','cur_company', 'school_index', 'location' })
    arr_ = [0] * 501
    arr_[1] = 1
    arr_[100] = 1
    arr_[201] = 1
    arr_[88] = 1
    arr_[34] = 1
    arr_[300] = 1
    
    mentee['skill_list'] = [arr_]
    mentee['cur_company'] = 'Amazon'
    mentee['location'] = 'Vietnam'
    mentee['school_index'] = [[0, 3]]
    
    return mentee
```
