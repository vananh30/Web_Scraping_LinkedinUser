```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import requests
```

```python
def login(browser):
    browser.get('https://www.linkedin.com/login?fromSignIn=true&trk=guest_homepage-basic_nav-header-signin')
    time.sleep(5)
    browser.find_element("name", "session_key").send_keys('' + Keys.RETURN)
    browser.find_element("name", "session_password").send_keys('' + Keys.RETURN)
    time.sleep(5)
```

```python
def user_dataframe_preprocessing(soup):
    name_tag = []
    job_tag = []
    location_tag = []
    profile_links = []
    title_tag = soup.find_all('div', class_='entity-result__item')
    for j in range(len(title_tag)):
        name = title_tag[j].find_all('span', class_= 'entity-result__title-text t-16')
        job = title_tag[j].find_all('div', class_= 'entity-result__primary-subtitle t-14 t-black t-normal')
        location = title_tag[j].find_all('div', class_= 'entity-result__secondary-subtitle t-14 t-normal')
        link = title_tag[j].find_all('a', href=True, class_ = 'app-aware-link scale-down')
        if len(name) != 0 and len(job) != 0 and len(location) != 0 and len(link) != 0: 
            name_tag.append(name[0].text.split('View')[0].strip())
            job_tag.append(job[0].text.strip())
            location_tag.append(location[0].text.strip())
            profile_links.append(link[0].get('href'))
    return  name_tag, job_tag, location_tag, profile_links 
```

```python
def load_jobs_dict(browser, jobDict):
    for i in range(len(jobDict.keys())):
        if i >= 12:
            connections_scraper(browser, list(jobTitleDict.keys())[i], jobDict[list(jobTitleDict.keys())[i]])
```

```python
# Import dataframe into MySQL
import sqlalchemy
def etl_mysql(connections):
    database_username = 'root'
    database_password = ''
    database_ip       = 'localhost'
    database_name     = 'LINKEDIN_USER'
    database_connection = sqlalchemy.create_engine('mysql+mysqlconnector://{0}:{1}@{2}/{3}'.
                                                   format(database_username, database_password, 
                                                          database_ip, database_name))
    connections.to_sql('users', database_connection, index=False, if_exists='append')
```

```python
def connections_scraper(browser, jobName, joblink):
    connections_page = joblink
    browser.get(connections_page)
    time.sleep(5)
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    time.sleep(3)
    i = 2
    while True:
        print('\nScraping your connections page: ', i, ' user doing ',jobName )
        name_tag, title_tag, location_tag, profile_links  = user_dataframe_preprocessing(soup)

        df1 = pd.DataFrame({'Name':name_tag, 'Title':title_tag, 'Location':location_tag, 'Profile':profile_links, 'JobName:': jobName})
        df1 = df1.drop(df1[df1.Name == 'LinkedIn Member'].index)
        etl_mysql(df1)
        print('\nSuccessfull loading page ', i, ' to mysql')
        browser.get(joblink + '&page='+str(i))
        time.sleep(5)
        if i == 100: 
            break    
        i+=1 
        time.sleep(10)
        soup = BeautifulSoup(browser.page_source, 'html.parser')
    return
```
